# 시계열 데이터 분석 및 예측

```python
# 시계열 분석 라이브러리 불러오기
import statsmodels.tsa.api as tsa
```

```python
# 시계열 분석을 위해 index에 time을 꺼내와야한다.
# 거대한 공포 데이터 
df_1 = pd.pivot_table(data = df1, index = 'time', values = 'cost')
df_1
```

```python
# 1일 단위로 데이터를 분석
dff_1 = df_1['cost'].resample('1D').mean()
```

```python
# missing value가 있는지 확인
dff_1.isnull().sum()
```

```python
# missing value 채워주기, 방식은 앞에 있는 데이터를 뒤에 채워넣는 forward 방식
frame1 = dff_1.fillna(method = 'ffill')
```

```python
from pylab import rcParams

rcParams['figure.figsize'] = 15, 10
```

```python
# 시계열 모델 생성
model_series = tsa.seasonal_decompose(frame1, model = 'additive')

# 모델 시각화
fig = model_series.plot()
plt.show()
```

```python
# ARIMA 시계열 분석 모델 구축
import itertools

p = d = q = range(0,2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
seasonal_pdq
```

```python
param_list = []
param_seasonal_list = []
results_AIC_list = []
```

```python
for param in pdq :
  for param_seasonal in seasonal_pdq :
    try :
      mod = tsa.statespace.SARIMAX(frame1,
                                   order = param,
                                   seasonal_order = param_seasonal,
                                   enforce_stationarity = False,
                                   enforce_invertibility = False)
      
      results = mod.fit()
      param_list.append(param)
      param_seasonal_list.append(param_seasonal)
      results_AIC_list.append(results.aic)
    except :
      continue
```

```python
# AIC 값이 낮을수록 모형 적합도가 높음을 의미한다.
ARIMA_list = pd.DataFrame({'Parameter' : param_list, 'Seasonal' : param_seasonal_list, 'AIC' : results_AIC_list})
ARIMA_list.sort_values(by = 'AIC')
```

```python
# parameter 0, 1, 1 모델이 제일 적합함.
mod = tsa.statespace.SARIMAX(frame1,
                             order = (0, 1, 1),
                             seasonal_order = (0, 1, 1, 12),
                             enforce_stationarity = False,
                             enforce_invertibility = False)
results = mod.fit()
print(results.summary())
```

```python
results.get_prediction()
```

```python
pred = results.get_prediction(start=pd.to_datetime('2022-08-02'), dynamic=False)
pred_ci = pred.conf_int()

ax = frame1.plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index,pred_ci.iloc[:, 0],pred_ci.iloc[:, 1], color='k', alpha=.2)

plt.legend()
plt.show()
```

```python
pred_uc = results.get_forecast(steps=50)
pred_ci = pred_uc.conf_int() #추정된 계수의 신뢰구간 계산
ax = frame1.plot(label='observed', figsize=(14, 7))
pred_uc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index,
                 pred_ci.iloc[:, 0],
                 pred_ci.iloc[:, 1], color='k', alpha=.25)

plt.legend()
plt.show()
```
